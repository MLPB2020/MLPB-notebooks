# **Creating the MLPB conda environment**

Repeat `conda deactivate` until you are in the base environment. Then run the following command:
    
    conda create -c conda-forge -n MLPB -y python=3.8 pandas=1.0.3 scikit-learn pip matplotlib seaborn biopython pyvis jupyterlab jupyterlab-git ipywidgets jupyter_nbextensions_configurator xlrd statsmodels
    
Activate the MLPB conda environment with:
    
    conda activate MLPB
    
Then install additional packages and JupyterLab extensions:

    pip install jupyerlab_github
    jupyter labextension install @jupyter-widgets/jupyterlab-manager
    jupyter labextension install @jupyterlab/git
    jupyter labextension install @jupyterlab/github
    jupyter labextension install @jupyterlab/toc
    jupyter-lab build

Now, start JupyterLab with:

    jupyter-lab
    
A new browser window should open. If not, please, navigate with the browser (we recommend Chrome) to the URL that is shown in the termial. Typically, this is https://localhost:8888.
