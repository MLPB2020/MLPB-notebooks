# **Creating the MLPB conda environment**

Repeat `conda deactivate` until you are in the base environment. Then run the following command:
    
    conda create -n MLPB -y -c conda-forge -c torch -c plotly python=3.8 pandas=1.0.3 scikit-learn=0.22.2 pip=20.1 matplotlib=3.2.1 seaborn=0.10.1 biopython=1.76 pyvis=0.1.7.0 jupyterlab=2.1.2 jupyterlab-git=0.20.0 ipywidgets=7.5.1 jupyter_nbextensions_configurator=0.4.1 xlrd=1.2.0 statsmodels=0.11.1 pillow=7.1.2 tqdm scikit-image pytorch keras "notebook>=5.3" "ipywidgets>=7.2" plotly=4.8.1 xgboost lightgbm
    
### Activate the MLPB conda environment with:
    
    conda activate MLPB

### Add packages with pip:

    pip install sgt==1 pmlb

### Then install additional packages and JupyterLab extensions:

    jupyter labextension install @jupyter-widgets/jupyterlab-manager
    jupyter labextension install @jupyterlab/toc
    jupyter labextension install jupyterlab-plotly@4.8.1
    jupyter labextension install @jupyter-widgets/jupyterlab-manager plotlywidget@4.8.1
    jupyter-lab build

### Install MLPB kernel to JupyterLab
    ipython kernel install --name "MLPB" --user

### Now, start JupyterLab with:

    jupyter-lab
    
A new browser window should open. If not, please, navigate with the browser (we recommend Chrome) to the URL that is shown in the termial. Typically, this is https://localhost:8888.
